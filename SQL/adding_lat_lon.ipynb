```python
import pgeocode 
import pandas as pd
import numpy as np
```

```python
csv = 'maintable.csv'
df = pd.read_csv(csv)
df
```

```python
# zipcode = df.loc[:,['zipcode']]
zipcode = pd.array(df['zipcode']).astype(str)
zipcode
```

```python
nomi = pgeocode.Nominatim('us')
# place = nomi.query_postal_code('19146')
# lat = place['latitude']
# print(lat)
latitudes = []
longitudes = []
for i in zipcode:
    place = nomi.query_postal_code(i)
    latitudes.append(place['latitude'])
    longitudes.append(place['longitude'])
df['latitudes'] = latitudes
df['longitudes'] = longitudes
df.to_csv('main_table_latslongs.csv')
```
